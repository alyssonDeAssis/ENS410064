```python
def corrige():
    if abs(z[1]-z[0])<d[0]*sin(radians(3.49)):
        z_corrigido[1]=z[1]
        theta[1]=abs(z_corrigido[1]-z_corrigido[0])
    else:
        z_corrigido[1]=z[0]+sign(z[1]-z[0])*d[0]*sin(radians(3.49))
        theta[1]=abs(z_corrigido[1]-z_corrigido[0])
    

    for i in range(2, len(z)):
        inclinacao=(z_corrigido[i-1]-z_corrigido[i-2])

        inclinacao_propagada=z_corrigido[i-1]+inclinacao
    
        if inclinacao >= 0.0:
            if z[i]>inclinacao_propagada and z[i]<inclinacao_propagada+d[i-1]*sin(radians(beta)):
                z_corrigido[i]=z[i]
            if z[i]>inclinacao_propagada and z[i]>inclinacao_propagada+d[i-1]*sin(radians(beta)):
                z_corrigido[i]=inclinacao_propagada+d[i-1]*sin(radians(beta))
            if z[i]<inclinacao_propagada and z[i]>inclinacao_propagada-d[i-1]*sin(radians(beta)):
                z_corrigido[i]=z[i]
            if z[i]<inclinacao_propagada and z[i]<inclinacao_propagada-d[i-1]*sin(radians(beta)):
                z_corrigido[i]=inclinacao_propagada-d[i-1]*sin(radians(beta))
        if inclinacao < 0:
            if z[i]<inclinacao_propagada and z[i]>inclinacao_propagada-d[i-1]*sin(radians(beta)):
                z_corrigido[i]=z[i]
            if z[i]<inclinacao_propagada and z[i]<inclinacao_propagada-d[i-1]*sin(radians(beta)):
                z_corrigido[i]=inclinacao_propagada-d[i-1]*sin(radians(beta))
            if z[i]>inclinacao_propagada and z[i]<inclinacao_propagada+d[i-1]*sin(radians(beta)):
                z_corrigido[i]=z[i]
            if z[i]>inclinacao_propagada and z[i]>inclinacao_propagada+d[i-1]*sin(radians(beta)):
                z_corrigido[i]=inclinacao_propagada+d[i-1]*sin(radians(beta))
        #theta[i]=abs(degrees(arcsin((abs(z_corrigido[i-1]-z_corrigido[i-2])-abs(z_corrigido[i]-z_corrigido[i-1]))/d)))
```

```python
def resultados(solution):

    for i in range(len(z)-1):
        z[i]=z_[i]+solution[i]
    
    #x_1=solution[0]
    #x_2=solution[1]
    #x_3=solution[2]
    #z[0]=z_[0]+x_1
    #z[(int(len(z)/2+1))]=z_[(int(len(z)/2+1))]+x_2
    #z[len(z)-1]=z_[len(z)-1]+x_3
    
    corrige()
    
    #theta[i]=abs(degrees(arcsin((abs(z_corrigido[i-1]-z_corrigido[i-2])-abs(z_corrigido[i]-z_corrigido[i-1]))/d[i])))
    
    return z_corrigido
```

```python
def remover_primeiro_ultimo(lista):
    """
    Remove o primeiro e o último elemento de uma lista.
    
    Parâmetros:
    lista (list): A lista original.
    
    Retorna:
    list: A lista sem o primeiro e o último elemento.
    """
    # Verificar se a lista tem mais de 2 elementos
    if len(lista) > 2:
        return lista[1:-1]
    else:
        return []  # Se tiver 2 ou menos elementos, retorna uma lista vazia
```

```python
def plotar_curva_deslocamento_forca(coord_x_estaca,z_,z_corrigido):
    """
    
    """
    
    # Primeira curva: terreno natural
    trace1 = go.Scatter(
        x=coord_x_estaca,
        y=z_,
        mode='lines+markers',  # Tipo de gráfico: linha e pontos
        name='terreno natural'
    )
    
    # Segunda curva: terreno alterado (exemplo de dados)
    trace2 = go.Scatter(
        x=coord_x_estaca,
        y=z_corrigido,
        mode='lines+markers',  # Tipo de gráfico: linha e pontos
        name='terreno alterado'
    )

    
    layout = go.Layout(
        title='Terraplenagem_trackers',
        xaxis=dict(title='Dist_horizontal'),
        yaxis=dict(title='z(m)'),
        showlegend=True
    )

    fig = go.Figure(data=[trace1, trace2], layout=layout)
    
    fig.update_layout(
        title=f'Terraplenagem de estacas',
        xaxis_title='Dist_horizontal',
        yaxis_title='z(m)',
        template="plotly_white"
    )
    
    # Exibe o gráfico
    pio.show(fig)

# Chamada da função para exibir o gráfico
#plotar_curva_deslocamento_forca()
```

```python
import pandas as pd

# Carregar os dados do arquivo .txt
arquivo_txt = 'C:\\Users\\alysson.assis\\Downloads\\INPUT_terraplenagem-Layout NOVO.txt'  # nome do arquivo de entrada
df= pd.read_csv(arquivo_txt, delimiter=',')
# Para cada tracker, realizar o cálculo e adicionar a coluna "ELEVACAO_CORRIGIDA"
trackers_unicos = df['TRACKER'].unique()

#print(trackers_unicos)

# Iterar sobre cada valor de TRACKER
for tracker in trackers_unicos:
    print(tracker)
    # Filtrar os dados do tracker atual
    tracker_data = df[df['TRACKER'] == tracker]
    z = tracker_data['ELEVACAO'].tolist()
    primeiro_z = z[0]
    ultimo_z = z[-1]
    #z = remover_primeiro_ultimo(z)

    norte = tracker_data['NORTE'].tolist()
    primeiro_norte = norte[0]
    ultimo_norte = norte[-1]
    #norte = remover_primeiro_ultimo(norte)

    d = [0] * (len(z)-1)
    #d[0] = abs(norte[1]-norte[0])
    for i in range(0,len(z)-1):
        d[i] = abs(norte[i]-norte[i-1])
    
    z_corrigido = [0] * len(z)
    theta = [0] * len(z)
    beta = 0.75

    z_corrigido[0] = z[0]

    corrige()
    
    z_ = tracker_data['ELEVACAO'].tolist()
    primeiro_z_ = z_[0]
    ultimo_z_ = z_[-1]
    #z_ = remover_primeiro_ultimo(z_)

    # define range for input
    r_min, r_max = -1.0, 1.0
    # define the bounds on the search
    bounds = [[r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max], [r_min, r_max]]
    # perform the differential evolution search
    result = differential_evolution(objective, bounds, maxiter=10000, popsize=15, mutation=(0.8, 1), recombination=0.9)
    # summarize the result
    print('Status : %s' % result['message'])
    print('Total Evaluations: %d' % result['nfev'])
    # evaluate solution
    solution = result['x']
    evaluation = objective(solution)
    print('Solution: f(%s) = %.5f' % (solution, evaluation))

    z_corrigido=resultados(solution)
    #z_corrigido = [primeiro_z_] + z_corrigido + [ultimo_z_]
    
    theta = [0] * len(z_corrigido)

    #d=[abs(norte[0]-primeiro_norte)] + d + [abs(ultimo_norte-norte[-1])]
    if d[0]==0:
        d[0]=1.49
    #print(d)
    #print(z_corrigido)
    #print('--')
    #print(z_)
    for i in range(1,len(z_corrigido)-1):
        theta[i]=abs(degrees(arcsin((abs(z_corrigido[i-1]-z_corrigido[i-2])-abs(z_corrigido[i]-z_corrigido[i-1]))/d[i])))
    #print(theta)

    
    # Adicionar a elevação corrigida ao DataFrame original
    df.loc[df['TRACKER'] == tracker, 'ELEVACAO_CORRIGIDA'] = z_corrigido

    coord_x_estaca = [0] * len(z_corrigido)
    coord_x_estaca[0] = df['NORTE'][0]
    
    for i in range(1,len(z_corrigido)):
        coord_x_estaca[i]=coord_x_estaca[i-1]-d[i-1]*cos(arcsin(radians(theta[i])))

    coord_x_estaca[-1]=coord_x_estaca[-2]-d[-1]
    
    df.loc[df['TRACKER'] == tracker, 'NORTE_CORRIGIDO'] = coord_x_estaca
    
    print("coord_x = ", coord_x_estaca)
    print("z_ = ", z_)
    print("z_corrigido =", z_corrigido)
    # Salvar dataframe atualizado
    df.to_csv(arquivo_txt, index=False)
```
