# Continuação tipos de variáveis

```python
# Exemplo lista
a = [1,2,3,10,20,30]

# Indexação da lista = a[0] =1
print(a[0])

# Acessando o último valor da lista
print(a[-1])
```

```python
# Determinando o comprimento da lista
b = len(a)

print("O comprimento da lista a é: ", b)

# Acessando o último valor da lista
print("O último valor da lista a é: ", a[b-1])
```

# O pacote numpy

https://numpy.org/

Este pacote serve para operações matriciais e estatísticas básicas.

```python
# Importando o numpy
import numpy as np

c= np.random.rand(100)
#print(c)
print(type(c))

# Acessando o último valor da lista
print("o último valor de c é: ", c[-1])

# Acessando os 10 últimos valores da lista
print("os dez primeiros números de c são: "+ str(c[0:10]))

# Loop com um array
for i in range(0,10):
    print(str(i) + ' ' +str(c[i]))

print(np.shape(c))
```

```python
# Criando uma matriz numpy

d = np.random.rand(100,100)
print(d)

print('Qual é o shape/dimensão da matriz?')
print(np.shape(d))

# Acessando a matriz numpy
# Linha 10 e coluna 10
print(d[10,10])
print(d[10][10])
```

```python
# Criando uma matriz numpy 3D

mat3D = np.random.rand(365,100,100)
print(np.shape(mat3D))
print(mat3D[49,49,49])
```

```python
# Acessando todos os valores da primeira dimensão na célula 49,49

timeSeries = mat3D[:,49,49]
print(timeSeries)

# Acessando os 10 últimos
timeSeries_ = mat3D[-10:,49,49]
print(timeSeries_)
```

```python
# Operação matriciais com numpy

timeSeries2 = mat3D[:,50:52,50]
print(np.shape(timeSeries2))
print(timeSeries2.shape)

timeSeriesTransposta = timeSeries2.transpose()
print(timeSeriesTransposta.shape)
```

```python
# Estatísticas com numpy

mediaTimeSeries = np.mean(timeSeries2)
print('Essa é a média de todos os valores')
print(mediaTimeSeries)

# Média nas colunas
mediaTScolumns = np.mean(timeSeries2, axis=0)
print(mediaTScolumns)

# Média nas linhas
mediaTSlines = np.mean(timeSeries2, axis=1)
#print(mediaTSlines)

# Média na terceira dimensão
media3D = np.mean(mat3D)
print(media3D)
print(media3D.shape)

# Mínimo na terceira dimensão
min3D = np.min(mat3D)
print(min3D)

# Máximo na terceira dimensão
max3D = np.max(mat3D)
print(max3D)

# Desvio na terceira dimensão
desvio3D = np.std(mat3D)
print(desvio3D)

# Percentil
pct3D = np.percentile(mat3D,90)
print(pct3D)

# Variância
var3D = np.var(mat3D)
var3D = desvio3D**2
print(var3D)

```

```python
# Contenação
mat3D2 = mat3D.copy()

matConcat = np.concatenate([mat3D,mat3D2])
print(matConcat.shape)
```

```python
# Concatenação em lista

listMat = [mat3D, mat3D, mat3D]

mat3Dconc = np.concatenate(listMat)
print(mat3Dconc.shape)
```

# Plotando com o matplotlib
* plot
* histograma
* bar
* boxplot
* pcolor
* scatter

```python
# Importando o pacote matplotlib

import matplotlib.pyplot as plt

# Criando um objeto matplotlib para plotar as figuras
fig,ax = plt.subplots(3)

# lineplot
ax[0].plot(mat3D[:,10,10])
ax[1].plot(mat3D[:,10:13,10])

# bar
ax[2].bar(np.array(range(0,mat3D.shape[0])),mat3D[:,10,10])
```

```python
# BOXPLOT

fig, ax = plt.subplots(2)

ax[0].boxplot(mat3D[:,10:20,0])
ax[1].boxplot(mat3D[:,10:20,0].flatten())
```

```python
# pcolor - gráfico de superfície

fig, ax = plt.subplots(2,2)

ax[0,0].pcolor(mat3D[0,:,:])#,cmap='jet') #jet, gray
ax[0,1].contour(np.sin(mat3D[0,:,:]),cmap='jet') #jet, gray
ax[1,1].scatter(np.sin(mat3D[0,:,:]),mat3D[0,:,:].flatten())
ax[1,0].hist(np.sin(mat3D[0,:,:].flatten()))
```
